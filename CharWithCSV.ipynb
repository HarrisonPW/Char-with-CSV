```python
def ChatWithPandas(message):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": """You are a helpful assistant that converts English sentence to SQL in postgresql.
We have the PLAYER table right now and there is the table schema:
CREATE TABLE IF NOT EXISTS PLAYER(
    sofifa_id INTEGER PRIMARY KEY,
    short_name TEXT,
    long_name TEXT,
    player_positions TEXT,
    overall INTEGER,
    potential INTEGER,
    wage_eur FLOAT(64,64),
    value_eur FLOAT(64,64),
    age INTEGER
); """},
                {"role": "user", "content": message},
            ],
        )

        sql_query = response.choices[0].message['content']

    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None

    return sql_query
```

```python
result_df = ChatWithPandas('output the sum of the age in the dataset')
print(result_df)
cursor.query(result_df).df()
```
